# PROFE AQUI ME FALLÓ HACER UN DICCIONARIO DE PALABRAS SIGUIENTES


```python
def palabras_siguientes(oraciones):
    #tokens = oraciones.split()
    #tokens = [palabra.lower() for palabra in tokens]
    
    types = []
    for palabra in oraciones:
        if palabra not in types:
            types += [palabra] 
    D = {}
    for palabra in types:
        D[palabra]=[]
        
    for i in range(len(oraciones)):
        palabra = oraciones[i]
        ## con algo parecido podrías buscar las palabras que están antes y después con una rango (+- 5 palabras)
        if i+1<len(oraciones):
            siguiente = oraciones[i+1]
            D[palabra]+=[siguiente]
    return D
```
